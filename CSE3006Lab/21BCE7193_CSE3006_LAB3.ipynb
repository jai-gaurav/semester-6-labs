Importing libraries

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot, column
from bokeh.models import ColumnDataSource, HoverTool, Slider, CustomJS 
from bokeh.sampledata.penguins import data as df
import seaborn as sns
import matplotlib.pyplot as plt
```

### Question 1

```python
df.head()
```

```python
source = ColumnDataSource(df)
```

First plot

```python
hover = HoverTool(tooltips=[("Species", "@species"), ("Bill Length (mm)", "@bill_length_mm"), ("Bill Depth (mm)", "@bill_depth_mm")])
plot1 = figure(title="Bill Length Vs Bill Depth (mm)", 
               tools=[hover, 'pan', 'box_zoom', 'wheel_zoom', 'reset'])
scatter1 = plot1.circle('bill_length_mm', 
                       'bill_depth_mm', 
                       source=source, size=5)
```

Second plot

```python
hover = HoverTool(tooltips=[("Species", "@species"), ("Bill Length (mm)", "@bill_length_mm"), ("Flipper Length (mm)", "@flipper_length_mm")])
plot2 = figure(title="Bill Length Vs Flipper Length (mm)", 
               tools=[hover, 'pan', 'box_zoom', 'wheel_zoom', 'reset'])
scatter2 = plot2.circle('bill_length_mm', 
                       'flipper_length_mm', 
                       source=source, size=5)
```

Make the ranges same

```python
plot1.x_range = plot2.x_range
```

Add slider for point size

```python
slider = Slider(start=1, end=10, value=5, step=0.5, title="Point Size")

callback = CustomJS(args=dict(source=source, scatter1=scatter1, scatter2=scatter2), 
                    code="""
                    scatter1.glyph.size = {'value': cb_obj.value};
                    scatter2.glyph.size = {'value': cb_obj.value};
                    source.change.emit();
                    """)

slider.js_on_change('value', callback)

layout = column(slider, gridplot([[plot1, plot2]]))
```

Show linked view

```python
show(layout)
```

Linked View for Cloud Multivariate Dataset: Visible VS IR Mean, Min and Max

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

sns.regplot(x='Visible_mean', y='IR_mean', data=df, ax=axes[0], scatter=False)
axes[0].set_title('Visible Mean vs IR Mean')

sns.regplot(x='Visible_min', y='IR_min', data=df, ax=axes[1], scatter=False)
axes[1].set_title('Visible Min vs IR Min')

sns.regplot(x='Visible_max', y='IR_max', data=df, ax=axes[2], scatter=False)
axes[2].set_title('Visible Max vs IR Max')

for ax in axes:
    ax.set_axisbelow(True)
    ax.grid(True)

plt.show()
```
