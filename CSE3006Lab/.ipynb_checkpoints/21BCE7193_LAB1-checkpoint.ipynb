Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

Import dataset

```python
df = pd.read_csv("penguins_size.csv")
df.head()
```

```python
df["island"].value_counts()
```

Bar Plot

```python
# Bar plot for Species
sns.countplot(x='species', data=df)
plt.title('Distribution of Penguin Species')
plt.show()
```

Pair Plot

```python
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables (Colored by Species)')
plt.show()
```

Box Plot

```python
# Box plot for Culmen Length across Species
sns.boxplot(x='species', y='culmen_length_mm', data=df)
plt.title('Box Plot of Culmen Length Across Penguin Species')
plt.show()
```

Correlation Heatmap

```python
correlation_matrix = df.drop(columns=['species', 'island', 'sex']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

Pie Chart

```python
species_distribution = df['species'].value_counts()
plt.pie(species_distribution, labels=species_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Penguin Species Distribution')
plt.show()
```

Histogram

```python
sns.histplot(df["culmen_length_mm"])
plt.title(f'Histogram of {"culmen_length_mm"}')
plt.xlabel("culmen_length_mm")
plt.ylabel('Frequency')
plt.show()
```

Violin Plot

```python
plt.figure(figsize=(8,6))
plt.title("Culmen Length Distribution")
sns.violinplot(data=df, x="species", y="culmen_length_mm")
plt.show()
```

Swarm Plot

```python
sns.swarmplot(x='island', y='culmen_depth_mm', data=df)
plt.title('Swarm Plot of Culmen Depth Across Penguin Islands')
plt.show()
```

Line Plot

```python
x = np.linspace(0, 10, 100)

fig=plt.figure()
plt.plot(x,np.sin(x-1.57))
plt.plot(x,np.cos(x))
plt.show()
```

Scatter Plot

```python
sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', hue='species', data=df)

plt.title('Scatter Plot of Culmen Length vs Culmen Depth (Grouped by Species)')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Depth (mm)')

plt.show()
```

Count plot

```python
sns.countplot(df["species"])
```

Joint Plot

```python

sns.jointplot(x='culmen_length_mm', y='culmen_depth_mm', data=df, kind='scatter')
plt.suptitle('Joint Plot of Culmen Length and Culmen Depth')
plt.show()
```

Density Plot

```python
sns.kdeplot(data=df, x='body_mass_g', hue='species', common_norm=False, fill=True)
plt.title('Density Plot of Body Mass Across Penguin Species')
plt.show()
```

Ridgeline Plot

```python
sns.kdeplot(data=df, x='culmen_length_mm', hue='species', fill=True, multiple='stack', common_norm=False)
plt.title('Ridgeline Plot of Culmen Length Across Penguin Species')
plt.show()
```

Hexbin Plot

```python
sns.jointplot(x='culmen_length_mm', y='culmen_depth_mm', data=df, kind='hex', marginal_kws=dict(bins=20))
plt.suptitle('Hexbin Plot of Culmen Length and Culmen Depth')
plt.show()
```

Donut Plot

```python
sex_distribution = df['sex'].value_counts()
plt.pie(sex_distribution, labels=sex_distribution.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.gca().add_artist(plt.Circle((0, 0), 0.2, color='white'))
plt.title('Donut Plot of Penguin Sex Distribution')
plt.show()
```

3D Scatter Plot

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.scatter(df['culmen_length_mm'], df['culmen_depth_mm'], df['flipper_length_mm'], c=df['body_mass_g'], cmap='viridis')
fig.colorbar(surface, label='Body Mass (g)')
ax.set_xlabel('Culmen Length (mm)')
ax.set_ylabel('Culmen Depth (mm)')
ax.set_zlabel('Flipper Length (mm)')
plt.title('3D Surface Plot of Culmen Dimensions')
plt.show()
```

Generate histogram for character count in your name

```python
def gen_hist(word: str):
    character_count = {}
    for character in word.lower():
        character_count[character] = character_count.get(character, 0) + 1
    
    characters, counts = zip(*sorted(character_count.items()))

    plt.bar(characters, counts, color='skyblue')
    plt.xlabel('Characters')
    plt.ylabel('Frequency')
    plt.title('Jai Gaurav 21BCE7194')
    plt.show()
```

```python
# word = input("Enter a word: ")
word = "Jai Gaurav 21BCE7193"
gen_hist(word)
```
